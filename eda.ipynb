```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load the Titanic dataset
df = pd.read_csv(r'C:\Users\katta\OneDrive\Desktop\titanic.csv')

# Display the first few rows of the dataset
print("First 5 rows of the dataset:")
print(df.head())

# 1. Summary statistics for numeric features
print("\nSummary statistics for numeric features:")
print(df.describe())

# 2. Check for missing values
print("\nMissing values in each column:")
print(df.isnull().sum())

# 3. Histograms for numeric features
# Plot histograms for numerical features like 'Age', 'Fare'
df[['Age', 'Fare']].hist(bins=30, figsize=(10, 5))
plt.suptitle('Histograms of Age and Fare')
plt.show()

# 4. Boxplots for numeric features
# Plot boxplots to identify outliers in 'Age' and 'Fare'
plt.figure(figsize=(10, 5))
sns.boxplot(data=df[['Age', 'Fare']])
plt.title('Boxplot of Age and Fare')
plt.show()

# 5. Pairplot to explore relationships between numerical features
# Plot pairplot to explore relationships between 'Age', 'Fare', and 'Pclass'
sns.pairplot(df[['Age', 'Fare', 'Pclass']])
plt.suptitle('Pairplot of Age, Fare, and Pclass', y=1.02)
plt.show()

# 6. Correlation matrix
# Plot a heatmap of the correlation matrix to see relationships between numerical features

# Only select numeric columns
correlation_matrix = df.select_dtypes(include=['number']).corr()

plt.figure(figsize=(10, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix')
plt.show()


# 7. Explore relationships with 'Survived' (target variable)
# Boxplot of 'Age' and 'Survived'
plt.figure(figsize=(10, 5))
sns.boxplot(x='Survived', y='Age', data=df)
plt.title('Boxplot of Age vs Survived')
plt.show()

# Count plot for 'Survived' (target variable) vs 'Pclass'
plt.figure(figsize=(10, 5))
sns.countplot(x='Pclass', hue='Survived', data=df)
plt.title('Countplot of Pclass vs Survived')
plt.show()

# 8. Feature-level inferences from visuals

# Age distribution comparison between survived and non-survived passengers
plt.figure(figsize=(10, 6))
sns.histplot(df[df['Survived'] == 0]['Age'], kde=True, color='red', label='Not Survived', stat='density', bins=30)
sns.histplot(df[df['Survived'] == 1]['Age'], kde=True, color='green', label='Survived', stat='density', bins=30)
plt.legend()
plt.title('Age Distribution: Survived vs Not Survived')
plt.show()

# Fare distribution comparison between survived and non-survived passengers
plt.figure(figsize=(10, 6))
sns.histplot(df[df['Survived'] == 0]['Fare'], kde=True, color='red', label='Not Survived', stat='density', bins=30)
sns.histplot(df[df['Survived'] == 1]['Fare'], kde=True, color='green', label='Survived', stat='density', bins=30)
plt.legend()
plt.title('Fare Distribution: Survived vs Not Survived')
plt.show()

# 9. Visualizing the relationship between categorical features and 'Survived'
# Count plot for 'Embarked' vs 'Survived'
plt.figure(figsize=(10, 5))
sns.countplot(x='Embarked', hue='Survived', data=df)
plt.title('Countplot of Embarked vs Survived')
plt.show()

# Count plot for 'Sex' vs 'Survived'
plt.figure(figsize=(10, 5))
sns.countplot(x='Sex', hue='Survived', data=df)
plt.title('Countplot of Sex vs Survived')
plt.show()

# 10. Heatmap of missing values to understand patterns of missingness
plt.figure(figsize=(10, 6))
sns.heatmap(df.isnull(), cbar=False, cmap='viridis')
plt.title('Heatmap of Missing Values')
plt.show()
```
